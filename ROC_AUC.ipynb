```python
'''
Understanding Roc Curves and AUC in Machine Learning with example

1. The Core Analogy: "The Security Guard":
    Imagine a security guard at an airport scanner. The scanner gives a "beep" score from 0 to 100.
    
    Low threshold: if gaurd stops everyone who beeps above 10, they will catch almost all terrorists(high recall), 
    but annoy thousands of innocent people(high false alarms)
    
    High threshold: If the guard only stops people who beep above 90, 
    they will never annoy an innocent person (zero false alarms), but they might let a terrorist walk through (low recall).

So this  curve can actually help us to find perfect threshold for the dataset

True Positive rate :
“Out of all actual positives, how many did the model correctly identify?”
-Formula:
 TPR = TP​ / TP + FN
 
 Example:
 if we have a confusion matrix like 
 80 | 20
 20 | 80 
 
 False negative is 20 and out of 100 actual positives, 80 were correctly identified. 
 and else where recognized as negatives (20). 
 So TPR is 80 / (80 + 20) = 0.8
 
 we aim to maximize the True positive rate 

False Positive rate:
“Out of all actual negatives, how many did the model incorrectly identify as positives?”
-Formula:
 FPR = FP / FP + TN
 So for our example we can say
 FPR is 20 / 20 + 80 = 0.2
 
 It is cost of model which is not ideal for any model. The lower the FPR, the better the model.
 

So idealy in Roc Curve TPR = 1 and FPR = 0,
and the curve is between this two values it can actually a graph between bensfit and cost of the model.

In Rocpng1 we can see that if threshold is 0, TPR is 1 and FPR is also 1, and if threshold is 1, TPR is 0 and FPR is also 0.
so the curve starts from (0,0) to (1,1) and there could be linear line but there is not a linear line


ROC and AUC curve stands for Receiver Operating Characteristic and Area Under the Curve respectively.
AUC is just area under the ROC curve from (0,0) and (1,1).
In practice ROC values lie between 0.5 to 1.0, and higher the AUC value, better the model performance.

 
'''
```
