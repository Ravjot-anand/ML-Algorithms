```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('student_clustering.csv')
print("The shape of data is",df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])

#here is data plotted in sctter plot
```

```python
from sklearn.cluster import KMeans
```

```python
#plotting the Elobw graph for the dataset to get what will be the value of K

wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
#from the graph it is visiblethat elebow point is 4 
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```
