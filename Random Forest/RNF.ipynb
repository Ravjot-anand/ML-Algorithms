```python
#Random Forest : is collection of decision tree and is a ensemble learning method
#It is a bagging technique if use decision tree as base learner

'''
Key words 
with replacement : rows get sampled with replacement and can appear multiple times in the same sample
without replacement : rows get sampled without replacement and can appear only once in the same sample
'''

'''
How does this algo performes so well?
So as we encounter the problem of low bias and high variance in decision trees
we use random forest to reduce the variance by averaging multiple decision trees(where each tree is trained on different subset of data,
and is fully grown without pruning)

It works so well as whole model comes out to be low variance not like individual decision trees which are high variance models
will see with the hepl of code example
'''
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
#Making toy dataset
np.random.seed(42)
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
#Seeing Decesion boundry of the decision tree
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train, y_train)

plt.figure(figsize=(12, 8))
x_range = np.linspace(X.min(), X.max(), 100)
xx1, xx2 = np.meshgrid(x_range, x_range)
y_hat = dtree.predict(np.c_[xx1.ravel(), xx2.ravel()])
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, xx2, y_hat, alpha=0.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha=.7)
plt.title("Decision tree")
plt.show()
```

```python
#Noticeable that decision tree is overfitting the data and has high variance
```

```python
#decision boundry of random forest
from sklearn.ensemble import RandomForestClassifier
     
rf = RandomForestClassifier(n_estimators=500, random_state=42)
rf.fit(X_train, y_train)

plt.figure(figsize=(12, 8))
x_range = np.linspace(X.min(), X.max(), 100)
xx1, xx2 = np.meshgrid(x_range, x_range)
y_hat = rf.predict(np.c_[xx1.ravel(), xx2.ravel()])
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, xx2, y_hat, alpha=0.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha=.7)
plt.title("Random forest")
plt.show()
```

```python
#Random Forest tries to capture the true essence of the data and is low variance
```
