```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv("IMDB_Dataset.csv")
```

We need to clean the data
1. Sample 10000 rows
2. Remove html tags
3. Remove special characters
4. Converting everything to lower case
5. Removing stop words(and, if, the, from, etc...)
6. Stemming(play, playing, played == play)


```python
#getting 10000 sample random rows
df = data.sample(10000)
```

```python
df.shape
```

```python
df.head()
```

```python
df['sentiment'].replace({'positive': 1, 'negative': 0}, inplace=True)
```

```python
df.head()
```

```python
#Removing Html Tags 

import re

def clean_HTML(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['review'] = df['review'].apply(clean_HTML)
```

```python
df['review'][0]
```

```python
#Converting text to lower case

def convert_Lower(text):
    return text.lower()
```

```python
df['review'] = df['review'].apply(convert_Lower)
```

```python
df['review'][0]
```

```python
#Removing special characters from the review

def remove_Special(text):
    x = ''
    
    for i in text:
        if i.isalnum():
            x = x + i
        else:
            x = x + ' '
    return x
```

```python
df['review'] = df['review'].apply(remove_Special)
```

```python
#Removing Stop words

import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_Stopwods(text):
    x = []
    
    for i in text.split():
        #this is to make a review in list
        if i not in stopwords.words('english'):
            #if there is no words from stpowords then append in empty x list
            x.append(i)
    
    # this next lines of code is to clear the x for next review
    y = x[:]
    x.clear()
    
    return y 
```

```python
df['review'] = df['review'].apply(remove_Stopwods)
```

```python
#Performing Stemming on the data

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

y = []

def stem_Word(text):
    for i in text:
        y.append(ps.stem(i))
    
    z = y[:]
    y.clear()
    
    return z
```

```python
df['review'] = df['review'].apply(stem_Word)
```

```python
df
```

```python
#Join Back  

def join_Back(list_input):
    return " ".join(list_input)
```

```python
df['review'] = df['review'].apply(join_Back)
```

```python
df['review']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
X = cv.fit_transform(df['review']).toarray()
```

```python
X.shape
```

```python
y = df.iloc[:,-1]
```

```python
y.shape
```

```python
y
```

Preparing data for training and testing


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

Training a model on data


```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
```

```python
Y_pred1 = clf1.predict(X_test)
Y_pred2 = clf2.predict(X_test)
Y_pred3 = clf3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('GaussianNB Accuracy:', accuracy_score(y_test,Y_pred1))
print('MultinomialNB Accuracy:', accuracy_score(y_test,Y_pred2))
print('BeronulliNB Accuracy:', accuracy_score(y_test,Y_pred3))
```
